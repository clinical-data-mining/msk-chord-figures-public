```python
def extractPercent(s):
    # extract from string ex: 170 (17%)
    if '%' in s:
        return float(s.split('%')[0].split('(')[1])/100
    return np.nan

def extractQuantiles(s):
    # extract from string ex: 36 (12,21,59,763)
    # returns median, 5%ile, 25%ile, 75%ile, 95%ile
    if ',' in s:
        tokens = re.split('\s|\(|\,|\)', s)
        return [float(t) for t in [tokens[0]]+tokens[2:-1]]
    return [np.nan]*5

def vizBoolSummary(ax,cbar_ax,summaryDf,color="#69d"):
    sns.heatmap(ax=ax,data=summaryDf.applymap(extractPercent),
               cmap=sns.light_palette(color, as_cmap=True),
               linewidths=1,cbar_ax=cbar_ax,xticklabels=False,
               annot=True, fmt='.1g',annot_kws={"fontsize":8})
    
def vizBoxRow(ax,summaryRow,log=False,color="#69d"):
    bxpstats = []
    for index, values in summaryRow.fillna('').items():
        statsumm = extractQuantiles(values)
        if log:
            penny=1
            statsumm = np.log10(np.array(statsumm)+penny)
        d = {'med':statsumm[0],
             'q1':statsumm[2],
             'q3':statsumm[3],
             'whislo':statsumm[1],
             'whishi':statsumm[4]}
        bxpstats+=[d]
        
    boxprops = dict(color=color)
    whiskerprops = dict(color=color)
    medianprops = dict(linewidth=1, color='white')

    ax.bxp(bxpstats,showfliers=False,showcaps=False,whiskerprops=whiskerprops,
           medianprops=medianprops,boxprops=boxprops,patch_artist=True)
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    name = summaryRow.name
    if log:
        name = 'log '+name
    ax.set_ylabel(name,rotation='horizontal',
                  labelpad=4,horizontalalignment='right',verticalalignment='center')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
   
def vizBarCategories(ax,summaryDf,miCols,miOther,ylabel,color,legendPos):
    
    percents = summaryDf.loc[miCols].fillna('').applymap(extractPercent)
    percents.loc[miOther] = 1-percents.sum()
    percents.loc[miOther,percents.columns[percents.isna().any(axis=0)]]=0
    percents = percents.fillna(0)

    groups = percents.columns
    values = percents.loc[miCols + [miOther]].values
    allCols = miCols + [miOther]

    for i in reversed(range(values.shape[0])):
        alpha = 1/(i+1) #0.5/(i+1)+0.5*(i%2)
        if values.shape[0]==2 and i==1 and not allCols[i]=='FEMALE':
            alpha=0.1
        ax.bar(groups, values[i], bottom = np.sum(values[:i], axis = 0),
              color=color,alpha=alpha, edgecolor='white', label=allCols[i],width = 0.9)

    # Stacked bar chart
    ax.set_xlim(-0.5,len(groups)-.5)
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel(ylabel,rotation='horizontal',
                  labelpad=4,horizontalalignment='right',verticalalignment='center')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.legend(loc='center left', bbox_to_anchor=(1+0.5*(legendPos%2), 0.5))
    
def vizSummary(summaryDf,miCols,otherCols,nameCols,filename,**kwargs):
    
    color='#69d'
    log=False
    if 'color' in kwargs:
        color=kwargs['color']
    if 'log' in kwargs:
        log=kwargs['log']
    
    
    summaryDf = summaryDf.fillna('')
    isBool = summaryDf.astype(str).apply(''.join,axis=1).str.contains('%')
    
    heatmapRows = []
    for r in summaryDf.index:
        if not r in list(isBool[~isBool].index)+[num for sublist in miCols for num in sublist]:
            heatmapRows+=[r]   

    fig,ax=plt.subplots(figsize=(1/5*len(summaryDf.columns),0.4*(sum(~isBool))+\
                        sum(0.3 + 0.1*np.array([len(a) for a in miCols]))+\
                               0.4*(sum(isBool)-len(otherCols))*(len(heatmapRows)>0)),
                        nrows=sum(~isBool) + len(otherCols) + (len(heatmapRows)>0),
                        ncols=2,
                        gridspec_kw={'height_ratios':[0.4]*sum(~isBool)+\
                                     list(0.3 + 0.1*np.array([len(a) for a in miCols]))+\
                                     [0.4*(sum(isBool)-len(otherCols))]*(len(heatmapRows)>0),
                                    'width_ratios':[0.9,0.1]})

    # draw boxplots
    for i, v in enumerate(isBool[~isBool].index): 
        vizBoxRow(ax[i,0],summaryDf.loc[v],log,color)
        ax[i,1].set_axis_off()
    
    # draw barplots
    for i in range(sum(~isBool),sum(~isBool)+len(otherCols)):
        vizBarCategories(ax[i,0],summaryDf,
                         miCols[i-sum(~isBool)],otherCols[i-sum(~isBool)],
                         nameCols[i-sum(~isBool)],color,i)
        ax[i,1].set_axis_off()         
    
    # draw heatmap
    if len(heatmapRows)>0:
        if sum(~isBool)+len(otherCols)>0:
            vizBoolSummary(ax[-1,0],ax[-1,1],summaryDf.loc[heatmapRows],color)
        else:  
            vizBoolSummary(ax[0],ax[1],summaryDf.loc[heatmapRows],color)
    plt.subplots_adjust(wspace=0, hspace=0)
    fig.savefig(filename,format=filename.split('.')[-1])
```

# Prostate (Gleason) visualizations

```python
def generic_volcano(x, y, genes, ax, adjust=True, **kwargs):
    x=pd.Series(x)
    y=pd.Series(y)
    genes=pd.Series(genes)
    padj = multipletests(y,alpha=0.05,method='fdr_bh')[1]
    
    #data['log2FoldChange'] = -np.log2(data['HR'])
    
    threshold = 0.05
    xns, yns = x[padj>=threshold], -np.log10(y[padj>=threshold])
    ax.scatter(xns, yns, c='grey', edgecolor=(1,1,1,0), label='Not Sig')
    xs, ys = x[padj<threshold], -np.log10(y[padj<threshold])
    ax.scatter(xs, ys, c='r', edgecolor=(1,1,1,0), label='FDR<5%')
    texts = []
    for xi, yi, l in zip(xs, ys, genes[padj<threshold]):
        texts.append(ax.text(xi, yi, l, size=8))
    ax.get_xaxis().get_major_formatter().labelOnlyBase = False
    ax.set_xlabel('$coeff$')
    ax.set_ylabel('$-log_{10}(pvalue)$')
    if adjust and len(texts)>0:
        adjust_text(texts, arrowprops=dict(arrowstyle="-", color='k', lw=0.5), ax=ax, **kwargs)
```

```python
from sklearn import linear_model
import statsmodels.api as sm

def gleasonCorr(df, genelist=common_genes): #change common_genes to full 341 gene list to reproduce main figure
    p = []
    coefs = []
    usedgenes = []
    for g in genelist:

        if sum(df[g])>=(len(df)*0.01):

            X = df['Gleason']
            y = df[g]

            reg = linear_model.LinearRegression()#fit_intercept = False)
            X2 = sm.add_constant(X)
            est = sm.OLS(y, X2)
            est2 = est.fit()
            p+=[est2.pvalues.Gleason]
            coefs+=[est2.params.Gleason]
            usedgenes+=[g]
    return p, coefs, usedgenes
```

```python
p, coefs, usedgenes = gleasonCorr(cancer2df_master_current_tx['prostate'])

fig,ax=plt.subplots(figsize=(4,4))
generic_volcano(coefs,p,usedgenes,ax)
ax.set_ylim(0,18)
ax.set_xlim(-0.02,0.08)
fig.savefig('figures/genes_gleason_volcano_small.pdf',format='pdf')
```

```python
# define some purer subsets of the original dataframes to study metastatic colonization
cancer2df_master_current_tx['LUAD'] = cancer2df_master_current_tx['nsclc'][cancer2df_master_current_tx['nsclc']['ADENOCARCINOMA'].astype(bool)]
cancer2df_master_current_tx['panc_adeno'] = cancer2df_master_current_tx['panc'][~cancer2df_master_current_tx['panc']['NONADENOCARCINOMA'].astype(bool)]
cancer2df_master_current_tx['HR+'] = cancer2df_master_current_tx['brca'][cancer2df_master_current_tx['brca']['HR'].astype(bool)]
cancer2df_master_current_tx['MSS_CRC'] = cancer2df_master_current_tx['crc'][~cancer2df_master_current_tx['crc']['MSI_OR_dMMR'].astype(bool)]
```

```python
def myHR(df_joint,var,**kwargs):
    duration_col='stop'
    event_col='dead'
    if 'duration_col' in kwargs:
        duration_col=kwargs['duration_col']
    if 'event_col' in kwargs:
        event_col=kwargs['event_col']
    df_joint = df_joint[df_joint[duration_col]>=0]
    if 'entry_col' in kwargs:
        cph = CoxPHFitter().fit(df_joint, entry_col=kwargs['entry_col'],
                                event_col=event_col, duration_col=duration_col, show_progress=False)
    else:
        cph = CoxPHFitter().fit(df_joint, event_col=event_col, duration_col=duration_col, show_progress=False)

    HR = (round(cph.hazard_ratios_[var], 2))
    lowerCI = (round(np.exp(cph.confidence_intervals_.loc[var,'95% lower-bound']),2))
    upperCI = (round(np.exp(cph.confidence_intervals_.loc[var,'95% upper-bound']),2))
    return (HR,lowerCI,upperCI,sum(df_joint[var]>0),len(df_joint),cph.summary.p)

def myHRDotplot(HRlist,namelist,threshold=0.05,join='inner',row_order=[]):
    HR = pd.DataFrame()
    for i, HRi in enumerate(HRlist):
        HRi=HRi.apply(lambda x: (np.nan,np.nan,np.nan,np.nan,[1]) if len(x)==0 else x)
        namei = namelist[i]
        if i==0:
            HR = pd.DataFrame(HRi.rename(namei))
        else:
            HR = pd.merge(HR,HRi.rename(namei),left_index=True, right_index=True, how=join)
        HR[namei] = HR[namei].apply(lambda x: (np.nan,np.nan,np.nan,np.nan,[1]) if x is np.nan else x)
        HR[namei+'_HR'] = HR[namei].apply(lambda x: x[0])
        HR[namei+'_N'] = HR[namei].apply(lambda x: x[3])
        HR[namei+'_pvalue'] = HR[namei].apply(lambda x: x[-1][0])
        HR[namei+'_padj'] = multipletests(HR[namei+'_pvalue'],alpha=threshold,method='fdr_bh')[1]
      
    for gblacklist in ['CDKN2B','FGF3','FGF4','FGF19']:
        if gblacklist in HR.index:
            HR = HR.drop([gblacklist])
            
    if len(row_order)>0:
        rows = []
        for r in row_order:
            if r in HR.index:
                rows+=[r]
        HR = HR.loc[rows]
    
    sigHR = HR.loc[HR[[namei+'_padj' for namei in namelist]].min(axis=1)<threshold]
    sig_hrs = sigHR[[namei+'_HR' for namei in namelist]]
    sig_padj = sigHR[[namei+'_padj' for namei in namelist]]
    sig_N = sigHR[[namei+'_N' for namei in namelist]]
    
    shape = sig_hrs.shape
    y = list(range(shape[0]))*shape[1]
    x = [item for sublist in [[i]*shape[0] for i in range(shape[1])] for item in sublist]
    colorlist = [sig_hrs.iloc[i[0],i[1]] for i in zip(y, x)]
    sizes = [8*sig_N.iloc[i[0],i[1]]**0.5 for i in zip(y, x)]
    padj_mask = [sig_padj.iloc[i[0],i[1]]<threshold for i in zip(y, x)]
    
    fig, ax = plt.subplots(figsize=(len(namelist)*3/4,len(sigHR)/2))

    divnorm=colors.TwoSlopeNorm(vmin=0., vcenter=1., vmax=3)
    if len(x)>0:
        plot = ax.scatter(x, y, s = sizes, c = colorlist, cmap = 'bwr', norm=divnorm)
        ax.scatter(pd.Series(x)[padj_mask],pd.Series(y)[padj_mask],s = pd.Series(sizes)[padj_mask],
                   edgecolors='k',facecolors='none')
        ax.set_xlim(-1,max(x)+1)
        ax.set_ylim(-1,max(y)+1)
        ax.set_xticks(list(range(shape[1])))
        ax.set_yticks(list(range(len(sigHR.index))))
        ax.set_xticklabels(namelist,rotation=90)
        ax.set_yticklabels(sigHR.index)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['left'].set_visible(False)
        fig.colorbar(plot,fraction=0.046)
    
    return fig,ax
```

```python
cancers = ['LUAD','HR+','MSS_CRC','prostate','panc_adeno']
sites = ['DMETS_DX_BRAIN','DMETS_DX_BONE','DMETS_DX_LIVER','DMETS_DX_LUNG']
control_cols = ['STAGE_IV_DX','ANY_PRIOR_TX']

# goal: fill a matrix HRs of tumor sites x cancers with series (of len # genes) 
#       corresponding to the HRs of the different genes for that site/cancer (adjusted for control_cols)
HRs = np.empty(shape=(len(sites),len(cancers)),dtype='object')

for j,cancer in enumerate(cancers):
    print(cancer)
    
    df_msk_current_tx = cancer2df_master_current_tx[cancer]
    
    for i,endpoint in enumerate(sites):
        print(endpoint)
        
        columns = control_cols + common_genes + ['stop_'+endpoint,'dead_'+endpoint]
        
        mask = (df_msk_current_tx['stop_'+endpoint]>=0)
        
        df_temp = df_msk_current_tx[mask][columns].fillna(0)
        HRs[i,j] = pd.Series()
        for g in common_genes:
            if len(df_temp[g].value_counts())>1 and (sum(df_temp[g])>(len(df_temp)*0.02)):
                HRs[i,j].loc[g] = myHR(df_temp[[g,'dead_'+endpoint,'stop_'+endpoint]+control_cols],g,
                                 event_col='dead_'+endpoint,duration_col='stop_'+endpoint)
```

```python
def flatten(l):
    return [item for sublist in l for item in sublist]
fig, ax = myHRDotplot(flatten(HRs.T),flatten([[c+s for s in sites]for c in cancers]),0.01,'outer')
fig.savefig('figures/metsites_dotplot_bpccommongenes.pdf',format='pdf')
```
